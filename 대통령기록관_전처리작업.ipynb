```python
import os
import pandas as pd
from tqdm import tqdm_notebook
path = 'Y:/crawler/faces/'
```

```python
file_list = os.listdir(path)

# .txt 파일 제거
file_list_dir = [x for x in file_list if not (x.startswith('bounding') or x.startswith('revision'))]

print("+로 시작하거나 끝나는 폴더 : {}".format([x for x in file_list_dir if x.endswith('+') or x.startswith('+')]))

meta = pd.read_excel('C:/Users/yeonsik/Desktop/200715_인물이미지_최종_이병기_중복_제거.xlsx')
meta['이름'] = [x.strip() for x in meta['이름']]
search_name = meta['비고.1'].dropna()

combine = [x for x in search_name if '+' in x or ',' in x]
for x in combine:
    print("'+' 나 ','가 있는 검색어 : {}".format(x))
```

** 위 리스트에서 아래 폴더명으로 이미지를 합침.

```python
from collections import Counter

search_name = list(search_name)

combine = [x for x in search_name if '+' in x or ',' in x]
print("'+' 기호나 ',' 기호가 있는 것들의 개수 : {}".format(len(combine)))

number_names = []
for i in tqdm_notebook(range(len(meta))):
    number_name = str(meta['연번'][i])+"_"+meta['이름'][i].strip().replace(' ','+')
    number_names.append(number_name)

meta['연번_이름'] = number_names

find_names = []
for i in tqdm_notebook(range(len(meta))):
    if str(meta['비고.1'][i]) == "nan":
        find_name = meta['이름'][i].strip().replace(' ','+')
    else:
        find_name = meta['비고.1'][i].strip().replace(' ','+')
    find_names.append(find_name)
meta['find_names'] = find_names

print("상위 3개의 해당 이름의 중복값이 있는지 확인용 {}".format(Counter(find_names).most_common(5)))
print("인물이미지 최종 개수 : {}".format(len(meta)))
```

```python
changed_names = []
for file in tqdm_notebook(file_list_dir):
    if file[-5:] == "_over":
        name = file[:-5]
        if name in find_names:
            name = list(meta[meta['find_names']==name]['연번_이름'])[0]+"_over"
        else:
            name = "find_me_"+name+"_over"
    else:
        name = file
        if name in find_names:
            name = list(meta[meta['find_names']==name]['연번_이름'])[0]
        else:
            name = "find_me_"+name
            
    changed_names.append(name)
```
